## 일일이 조사 결과
- 절도 == 절도범죄
- 강간강제추행: 강제추행  +  강간
- 유사강간는 12년도는 없음
- 기타강간강제추행등은 13년도는 기타강간강제추행으로 이름이 있거나 12년도에는 없다.

```python
# 같은 지역의 도시를 합처 지역으로만 나눠야 겠다.
# 함수로 데이터 처리
# 12 ~ 14년까지 도시로 나누어진 컬럼을 행정구역으로 나누는 함수
def indexprocess(dfC):
    df = dfC.copy()
    for i in df.index:
        df.iloc[i , 1] = df.iloc[i , 1].replace('절도범죄', '절도')
        df.iloc[i , 1] = df.iloc[i , 1].replace('기타강간강제추행', '기타강간강제추행등')
        df.iloc[i , 1] = df.iloc[i , 1].replace('기타강간강제추행등등', '기타강간강제추행등')
    df.index = df['범죄중분류']
    if '강간강제추행' not in df.index:
        df.loc['강간강제추행'] = df.loc['강간' ,:] + df.loc['강제추행',:]
        df = df.drop(index=['강간','강제추행'])
        df.loc['강간강제추행' , '범죄대분류'] = '강력범죄'
        df.loc['강간강제추행' , '범죄중분류'] = '강간강제추행'

    if '유사강간' not in df.index:
        df.loc['유사강간'] = np.nan
        df.loc['유사강간' , '범죄대분류'] = '강력범죄'
        df.loc['유사강간' , '범죄중분류'] = '유사강간'

    if '기타강간강제추행등' not in df.index:
        df.loc['기타강간강제추행등'] = np.nan
        df.loc['기타강간강제추행등' , '범죄대분류'] = '강력범죄'
        df.loc['기타강간강제추행등' , '범죄중분류'] = '기타강간강제추행등'

    df = df.sort_index()
    df.index = range(len(df))
    return df

def city_12_14_process(dfC):
    dfh = dfC.copy()
    df = indexprocess(dfh)

    dfName = pd.DataFrame()
    if '마산' not in df.columns:
        df['마산'] = np.nan # 2014년 마산이 없음
    df.columns = df.columns.str.strip()
    dfName['범죄대분류'] = df['범죄대분류']
    dfName['범죄중분류'] = df['범죄중분류']
    dfName['서울'] = df.loc[: , df.columns.str.startswith('서울')].sum(axis=1)
    dfName['경기'] = df.loc[: , ['수원', '성남', '부천', '안양', '고양', '의정부', '안산', '광명', '군포', '남양주', '평택', '시흥', '파주', '이천', '구리', '용인']].sum(axis=1)
    dfName['인천'] = df.loc[: , ['인천']].sum(axis=1)
    dfName['부산'] = df.loc[: , ['부산']].sum(axis=1)
    dfName['대구'] = df.loc[: , ['대구']].sum(axis=1)
    dfName['광주'] = df.loc[: , ['광주']].sum(axis=1)
    dfName['대전'] = df.loc[: , ['대전']].sum(axis=1)
    dfName['울산'] = df.loc[: , ['울산']].sum(axis=1)
    dfName['세종'] = np.nan
    dfName['강원'] = df.loc[: , ['춘천', '원주', '강릉']].sum(axis=1)
    dfName['충북'] = df.loc[: , ['청주', '충주', '제천']].sum(axis=1)
    dfName['충남'] = df.loc[: , ['천안', '아산', '논산', '서산']].sum(axis=1)
    dfName['전북'] = df.loc[: , ['전주', '익산', '군산', '정읍']].sum(axis=1)
    dfName['전남'] = df.loc[: , ['여수', '목포', '순천']].sum(axis=1)
    dfName['경북'] = df.loc[: , ['포항', '구미', '안동', '경산', '경주', '김천']].sum(axis=1)
    dfName['경남'] = df.loc[: , ['창원', '진주', '김해', '양산', '거제', '마산']].sum(axis=1)
    dfName['제주'] = df.loc[: , ['제주']].sum(axis=1)
    dfName['기타도시'] = df.loc[: , df.columns.str.startswith('기타도시')].sum(axis=1)
    dfName['도시이외'] = df.loc[: , df.columns.str.startswith('도시이외')].sum(axis=1)

    return dfName

# 15 ~ 23년까지 컬럼은 행정구역의 이름이 포함되여서 이 함수로 행정구역으로 나누는 함수
def dfprocess(dfC):
    dfh = dfC.copy()
    df = indexprocess(dfh)
    dfName = pd.DataFrame()
    dfName['범죄대분류'] = df['범죄대분류']
    dfName['범죄중분류'] = df['범죄중분류']
    dfName['서울'] = df.loc[: , df.columns.str.startswith('서울')].sum(axis=1)
    dfName['경기'] = df.loc[: , df.columns.str.startswith('경기')].sum(axis=1)
    dfName['인천'] = df.loc[: , df.columns.str.startswith('인천')].sum(axis=1)
    dfName['부산'] = df.loc[: , df.columns.str.startswith('부산')].sum(axis=1)
    dfName['대구'] = df.loc[: , df.columns.str.startswith('대구')].sum(axis=1)
    dfName['광주'] = df.loc[: , df.columns.str.startswith('광주')].sum(axis=1)
    dfName['대전'] = df.loc[: , df.columns.str.startswith('대전')].sum(axis=1)
    dfName['울산'] = df.loc[: , df.columns.str.startswith('울산')].sum(axis=1)
    dfName['세종'] = df.loc[: , df.columns.str.startswith('세종')].sum(axis=1)
    dfName['강원'] = df.loc[: , df.columns.str.startswith('강원')].sum(axis=1)
    dfName['충북'] = df.loc[: , df.columns.str.startswith('충북')].sum(axis=1)
    dfName['충남'] = df.loc[: , df.columns.str.startswith('충남')].sum(axis=1)
    dfName['전북'] = df.loc[: , df.columns.str.startswith('전북')].sum(axis=1)
    dfName['전남'] = df.loc[: , df.columns.str.startswith('전남')].sum(axis=1)
    dfName['경북'] = df.loc[: , df.columns.str.startswith('경북')].sum(axis=1)
    dfName['경남'] = df.loc[: , df.columns.str.startswith('경남')].sum(axis=1)
    dfName['제주'] = df.loc[: , df.columns.str.startswith('제주')].sum(axis=1)
    dfName['기타도시'] = df.loc[: , df.columns.str.startswith('기타도시')].sum(axis=1)
    dfName['도시이외'] = df.loc[: , df.columns.str.startswith('도시이외')].sum(axis=1)


    return dfName

df12 = city_12_14_process(df012)
df13 = city_12_14_process(df013)
df14 = city_12_14_process(df014)
df15 = dfprocess(df015)
df16 = dfprocess(df016)
df17 = dfprocess(df017)
df18 = dfprocess(df018)
df19 = dfprocess(df019)
df20 = dfprocess(df020)
df21 = dfprocess(df021)
df22 = dfprocess(df022)
df23 = dfprocess(df023)
```

```python
import copy
# 리스트 안에 데이터들이 복사하기 위해 가져옴
```

```python
# 깊은 복사로 데이터 원본 데이터 반영 안하기 위해서
alldf = copy.deepcopy([df12,df13,df14,df15,df16,df17,df18,df19,df20,df21,df22,df23])

for i in alldf:
    i['지역 합계'] = i.select_dtypes(np.number).sum(axis=1 , skipna=True)
topcrime =list(reduce(lambda x,y: x&y ,(set(i.sort_values(by='지역 합계' ,ascending=False)['범죄중분류'][:5]) for i in  alldf)))
topcrime
```

```python
# 범죄량 기준으로 top5와 그외 모든범죄를 내림차순으로 나타낸다.
dffff = pd.DataFrame()
for i in range(len(alldf)):
    dffffg = pd.DataFrame()
    dfg = alldf[i].copy()
    alldf[i]['지역 합계'] = alldf[i].select_dtypes(np.number).sum(axis=1 , skipna=True)

    a = str(i+12)+' 대표 탑 6'
    dfg.loc['범죄 합계'] = dfg.select_dtypes(np.number).sum(axis=1 , skipna=True)
    dffffg[a] = dfg[dfg['범죄중분류'].isin(topcrime)].sort_values(by='지역 합계' ,ascending=False)['지역 합계']
    b = str(i+12)+'범죄중분류'
    dffffg[b] = dfg[dfg['범죄중분류'].isin(topcrime)].sort_values(by='지역 합계' ,ascending=False)['범죄중분류']

    dfsg =  dfg[~dfg['범죄중분류'].isin(topcrime)]['지역 합계'].sum()

    dffffg.loc[len(dffffg) +1] = [dfsg, '그외모든범죄']

    dffffg = dffffg.sort_values(by=a ,ascending=False)

    dffffg.index = range(len(dffffg))
    dffff = pd.concat([dffff,dffffg],axis=1)
dttn = dffff.select_dtypes(np.number)
dstn = dffff.select_dtypes(include='object')
dffff
```
